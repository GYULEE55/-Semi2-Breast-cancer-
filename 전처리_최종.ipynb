##Income(연속형 그대로 사용)

```python
df['Income'].value_counts()
```

```python
# 1. Unknown 제거
df = df[df['Income'] != 'Unknown/missing/no match/Not 1990-2023']
```

```python

# 2. $와 , 제거 (raw string 사용)
df['Income_clean'] = df['Income'].str.replace(r'[\$,]', '', regex=True)
```

```python
# 예: 안전한 숫자 최소값 추출 함수 (사용자 환경에 맞게 이미 정의된 것으로 가정)
def extract_min_income_safe(income_str):
    if pd.isna(income_str):
        return None
    try:
        income_str = income_str.replace(',', '').strip()
        if '-' in income_str:
            low, _ = income_str.split('-')
            val = int(low)
        else:
            val = int(income_str)
        return val
    except:
        return None

# 1. Income_clean 컬럼이 있다고 가정하여 진행 (만약 없으면 'Income'으로 교체)
df['Income_num'] = df['Income_clean'].apply(extract_min_income_safe)

# 2. 그룹핑 함수
def income_group(x):
    if x is None:
        return None
    elif x < 55000:
        return 'LOW'
    elif 55000 <= x <= 89999:
        return 'MIDDLE'
    else:
        return 'HIGH'

# 3. Income_num 기준 그룹 생성
df['Income_group'] = df['Income_num'].apply(income_group)

# 4. 결측 제거 (필요시)
df = df[df['Income_group'].notna()]

# 5. 원핫 인코딩 생성
income_dummies = pd.get_dummies(df['Income_group'], prefix='Income').astype(int)

# 6. df_processed 인덱스 동기화 및 Income_group, 원핫인코딩 병합
df_processed = df_processed.reindex(df.index)
df_processed['Income_group'] = df.loc[df_processed.index, 'Income_group']
df_processed = pd.concat([df_processed, income_dummies], axis=1)

# 7. 결과 확인
print(df[['Income', 'Income_clean', 'Income_num', 'Income_group']].head())
print(income_dummies.head())
```

##Cancer treatment delay



```python
df['cancer treatment delay'].value_counts()
```

```python
import numpy as np

def treatment_delay_fine_group(value):
    if pd.isna(value) or value == 'Unable to calculate':
        return np.nan
    if value in ['0', 0, '0-14일']:
        return 0
    elif value in ['1', 1, '15-30일']:
        return 1
    elif value in ['2', 2, '31-60일']:
        return 2
    elif value in ['3', '4', 3, 4, '61-90일', '91일 이상']:
        return 3
    else:
        try:
            day = int(str(value).replace('일', '').replace(' ', '').replace('~','').replace('-',''))
            if day <= 14:
                return 0
            elif day <= 30:
                return 1
            elif day <= 60:
                return 2
            else:
                return 3
        except:
            return np.nan

df_processed['treat_delay_fine_group'] = df['cancer treatment delay'].apply(treatment_delay_fine_group)
df_processed = df_processed[df_processed['treat_delay_fine_group'].notna()]

print(df_processed['treat_delay_fine_group'].value_counts(dropna=False))
```
